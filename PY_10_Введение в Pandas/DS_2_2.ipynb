```python
import pandas as pd

def create_medications(names, counts):
    medications = pd.Series(data = counts, index = names)
    return medications

def get_percent(medications, name):    
    return medications.loc[name]*100/sum(medications)
    

names = ['chlorhexidine', 'cyntomycin', 'afobazol']
counts = [15, 18, 7]

medications = create_medications(names, counts)

print(create_medications(names, counts))
print(get_percent(medications, "chlorhexidine"))
```

```python
import pandas as pd

def create_companyDF(income, expenses, years):
    create_companyDF = pd.DataFrame({
    'Income': income,
    'Expenses': expenses,    
})
    create_companyDF.index = years 
    return create_companyDF
    

def get_profit(df, year):
    if year in df.index:
        return df.loc[year, 'Income']-df.loc[year, 'Expenses']
    else:
        return None
   


income = [478, 512, 196]
expenses = [156, 130, 270]
years = [2018, 2019, 2020]

df = create_companyDF(income, expenses, years)

print(create_companyDF(income, expenses, years))
print(get_profit(df, 2019))
print(get_profit(df, 2023))
```
